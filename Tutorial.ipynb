# Working with Image Features 

This segment works with:
- Size
- Color
- Resolution and neatness
- Rotation and Cropping an image
- Pasting and image into another image

We will modify the size and the output image into multiple sizes and change the dimensions of the image.  

Resizing images can result very useful; particularly, when storing significant amounts of images into a database, or when extracting them from a server. making them into a smaller and uniform size in order to fit them into a dataset. 

```python
# specify the size for the image
myimage2 = Image.open("Images/Parrot.jpg")

# Image original size, in pixels. The size is given as a 2-tuple (width, height).
print("Original Image Size: ", myimage2.size) 

# we set the size of the thumbnail we want to create out of the original image
size250 = (250,250)

# converts the original image into a thumbnail of the size specified above
myimage2.thumbnail(size250)
myimage2.save("Images/image4.jpg")

# The pixel format used by the image. Typical values are 1, L, RGB, or CMYK.
print("Image Mode: ", myimage2.mode) 

# Color palette table, if any.
print("The palette for this image: ", myimage2.palette) 

# To change the color to black and white
myimage2.convert(mode ='L').save("Images/image5.jpg")

# To make a blurry image 
# The default value is set to 2 which does not make much of a difference 
# but if incremented you will see a notorious difference in the clarity of the image
myimage2.filter(ImageFilter.GaussianBlur(15)).save("Images/BlurryParrot.jpg")

#rotate the image
myimage2.rotate(45).save("Images/ParrotRotated45.jpg")

#cropping and image
box = (95, 95, 180, 180)                         #---> First we create the box that will select the area we want to crop  
cropped_image = myimage2.crop(box)               #---> Crops the section we want
cropped_image.save("Images/cropped_Parrot.jpg")  #---> Saves the cropped image
```

#### Pasting an Image into Another Image   

This is usually used in advertisement where a logo is within the image. Also,on images that used watermarks on them, and in general where images are merged together.

From our Images folder, we load two images "cockatoo" being the main image and "moon-logo" as the logo  
 
 >[Images/Cockatoo.jpg](https://github.com/wvelebanks/Tutorial-03-18/blob/master/Tutorial/Images/Cockatoo.jpg) and 
 [Images/moon-logo.jpg](https://github.com/wvelebanks/Tutorial-03-18/blob/master/Tutorial/Images/moon-logo.jpg)

Below, myimage3 is the background image or the bigger image in the final output, logo is the image that will serve as the signature image contained in the background image.

we then create a copy of the background image in order to generate the dimmensions for the position of the logo in the resulting output. Then we paste the logo in the position previously calculated with the background image and finally we save the result of the two images merged.  

```python
# Opens the image that will serve as the main/background image  
myimage3 = Image.open("Images/Cockatoo.jpg")

#Opens the image that will serve as the logo/smaller image
logo = Image.open("Images/moon-logo.jpg")

#Created the image to be copied
image_copy = myimage3.copy()

# we set the position of the image within the bigger image 
position = ((image_copy.width - logo.width), (image_copy.height - logo.height))

# and finally we paste the smaller image  (Foreground) into the bigger image (background) 
image_copy.paste(logo, position)

# Saves the new merged image
image_copy.save("Images/pasted_image.jpg")
```

#### Changing Image Tonalities

- <strong> Embossing </strong> on which we hihglight and replace the boundaries with low contrast, and set a gray background on the image. this teachnique calculates the first derivative of the image and combines it with a uniform grayscale.


- <strong>Edge Detection</strong> works by detecting discontinuities in brightness; Edge detection is used for image segmentation and data extraction in areas such as image processing, computer vision, and machine vision.


- <strong>Colorization</strong> Colorization is commonly used in image processing  particularly to lower image resolution  to colorize a black-and-white photo, space missions relied upon this technique to keep data volumes low.


- <strong>Solarization</strong> is a technique which can be applied to any negative in the printing process.

```python
# Opens the Cockatoo image file 
myimage9 = Image.open("Images/Cockatoo.jpg")

#creates a image with the "Emboss" filter in it 
emboss_filter = myimage9.filter(ImageFilter.EMBOSS)

#Shows the image in the photo viewer
emboss_filter.show()

#saves the image in the folder specified
emboss_filter.save("Images/emboss.jpg")
```

```python
# applies the "find_edges" filter                                       
find_edges = myimage9.filter(ImageFilter.FIND_EDGES)

# display the output image in the photo viewer
find_edges.show()

# saves the output image in the folder Images
find_edges.save("Images/find_edges.jpg")
```

```python
# Opens a gray scale image of the Cockatoo
myimage10 = Image.open("Images/Cockatoo.jpg").convert("L")  

# Changes the color of the image into a bluish color by using the colorize filter 
myimage11 = ImageOps.colorize( myimage10, black="midnightblue", white="white")

#Save the resulting image with the new color
myimage11.save("Images/ColorizedImage.jpg")

# display the output image in the photo viewer
myimage11.show()
```

Image result from the photo viewer after applying the <strong>_Colorized_</strong> filter:

![Colorized Filter](Tutorial/Images/ColorizedImage.png)  

```python
# Opens a gray scale image of the Cockatoo
myimage10 = Image.open("Images/Cockatoo.jpg")      

# the first we apply the default value in the threshold parameter to solarize the image
result = ImageOps.solarize(myimage10, threshold=128)  

# saves image with 128 threshold
result.save("Images/solarizeImage1.jpg")

# then  we changed the values for the threshold parameter to 50 
result = ImageOps.solarize(myimage10, threshold=50)

# saves image with 50 threshold
result.save("Images/solarizeImage2.jpg")

# finally  we changed the values for the threshold parameter to 0 which is the same as using ImageOps.invert filter  
result = ImageOps.solarize(myimage10, threshold=0)  

# saves image with 0 threshold
result.save("Images/solarizeImage3.jpg")

# Show the last resulted image
result.show()
```

# Manipulating Image Pixels

In this last section we will work with more detailed ways to manipulate an image through its pixels. We will combine some of functions previously showcased to manipulate the image.

Another library useful for this purpose is <strong> _Mathplotlib_</strong> and <strong> _pylab_ </strong> because they both allow you to manipulate the numerical data contained in the pixels through the use of arrays, and displaying those values in a form of graph as you will see below with the examples one with a histogram equalization of an image, the change of colors by manipulating the array, and finally drawing lines inside an image with pixel dimensions. 

```python
# Image Equalization
# the first line opens the original image 
myimage13 = Image.open("Images/racoon.jpg")

# saves the image in black and white (gray scale)
myimage14 = myimage13.convert("L")  # grayscale

# saves the grayscale image to be equalized
myimage14.save("Images/equalizeImage.jpg")

# Equalize the image histogram.
myimage15 = ImageOps.equalize(myimage13)

#saves the first equalization/ (adds more contrast to the image)
myimage15.save("Images/equalizedhist.jpg")

#does the grayscale conversio again of the already equalized image 
myimage16 = myimage15.convert("L") 

#saves the second equalization of the image
myimage16.save("Images/equalizeRacoon.jpg")

# created and visualize the histogram derived from the different equalizations 
# performed in the image
bins = list(range(256))
plt.plot(bins, myimage14.histogram(), 'r')
plt.plot(bins, myimage16.histogram(), 'g')
plt.xlabel('Pixel value')
plt.ylabel('Frequency')
plt.title("Racoon Image Equalization")
plt.legend(
    ('src (grayscale)', 'dest (grayscale)'),
     shadow=True, loc=(0.01, 0.75))
plt.grid(True)
plt.savefig("Images/equalize_hist_01.jpg")
```
